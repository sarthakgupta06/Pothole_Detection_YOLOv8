#**Initialize the Model**

```python
model = YOLO('yolov8n.pt')  # 'yolov8n.pt' is the smallest version, ideal for limited resources
```

#**Train the Model**

```python
model.train(
    data='/content/drive/MyDrive/Pothole_Detection_System/data.yaml',  # Path to your data.yaml file
    epochs=100,  # Adjust epochs as needed
    batch=16,    # Batch size
    imgsz=640,   # Image size (640 is typical)
    name='pothole_detection_yolov8'  # Project name for the run
)
```

#**Evaluating The Model**

```python
metrics = model.val()  # This will output performance metrics
```

```python
# Directly call metrics to see the summary output
print(metrics)
```

#**Testing the Model**

```python
results = model.predict(source='/content/drive/MyDrive/sample_pothole.jpg', conf=0.5)
```
